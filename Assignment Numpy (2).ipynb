```python
#19. Question: Create a random 4x4 matrix and use NumPy to compute itseigenvalues and eigenvectors. Verify the eigenvalues by reconstructing the matrix.

A = np.random.rand(4,4)
values, vectors = np.linalg.eig(A)

reconstructed = vectors @ np.diag(values) @ np.linalg.inv(vectors)
print(reconstructed)
```

```python
#  1) Bar chart showing the total deaths per year for 1984-2016

df_84_16 = df[(df["year"] >= 1984) & (df["year"] <= 2016)]
yearly_deaths = df_84_16["year"].value_counts().sort_index()

plt.figure(figsize=(12,5))
yearly_deaths.plot(kind="bar")
plt.title("Total Police Deaths per Year (19842016)")
plt.xlabel("Year")
plt.ylabel("Total Deaths")
plt.show()
```

```python
# 2) Line chart comparing the yearly deaths in different states
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

top_states = df["state"].value_counts().head(5).index

state_year_counts = (
    df[df["state"].isin(top_states)]
    .groupby(["year", "state"])
    .size()
    .reset_index(name="deaths"))


plt.figure(figsize=(12,6))
sns.lineplot(
    data=state_year_counts,
    x="year",
    y="deaths",
    hue="state")

plt.title("Yearly Police Deaths in Top States")
plt.xlabel("Year")
plt.ylabel("Number of Deaths")
plt.show()
```

```python
# 4) A Choropleth comparing the number of police deaths in different states
import geopandas as gpd
state_deaths = df.groupby("state").size().reset_index(name="deaths")

states_map = gpd.read_file(
    "https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json")

states_map["state"] = states_map["id"]

choropleth_df = states_map.merge(state_deaths, on="state", how="left")
choropleth_df["deaths"] = choropleth_df["deaths"].fillna(0)

plt.figure(figsize=(15,10))

choropleth_df.plot(
    column="deaths",
    cmap="OrRd",
    linewidth=0.8,
    edgecolor="black",
    legend=True
)

plt.title("Choropleth Map of Police Deaths by State", fontsize=16)
plt.axis("off")
plt.show()
```

```python
# 5) A word cloud of the different causes of death (remove the string "Cause of Death:" for best results)
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

causes = df["cause"].dropna()

clean_causes = causes.str.replace(
    "Cause of Death:", "", regex=False)

text = " ".join(clean_causes)

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    colormap="viridis"
).generate(text)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Causes of Police Deaths", fontsize=14)
plt.show()
```

```python
# 6) A marker cluster showing the shootings in the state of California (show person's name on hover)

import pandas as pd
import folium
from folium.plugins import MarkerCluster


ca_shootings = df[
    (df["state"] == "CA") &
    (df["cause_short"].str.contains("Gunfire", na=False))]

california_map = folium.Map(
    location=[36.7783, -119.4179],  # Center of California
    zoom_start=6)

marker_cluster = MarkerCluster().add_to(california_map)

for _, row in ca_shootings.iterrows():
    folium.Marker(
        location=[36.7783, -119.4179],
        popup=f"Name: {row['person']}",
        tooltip=row["person"]
    ).add_to(marker_cluster)

california_map
```

```python
# 7) Bar chart comparing the no. of deaths due to different causes, animated by year

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation


df["cause_clean"] = df["cause"].str.replace(
    "Cause of Death:", "", regex=False
)

year_cause = (
    df.groupby(["year", "cause_clean"])
      .size()
      .reset_index(name="deaths")
)

years = sorted(year_cause["year"].unique())

fig, ax = plt.subplots(figsize=(10,6))

def update(year):
    ax.clear()
    data = year_cause[year_cause["year"] == year]
    ax.bar(data["cause_clean"], data["deaths"])
    ax.set_title(f"Police Deaths by Cause  Year {year}")
    ax.set_xlabel("Cause of Death")
    ax.set_ylabel("Number of Deaths")
    ax.tick_params(axis='x', rotation=45)

# IMPORTANT: keep animation in a variable
anim = FuncAnimation(fig, update, frames=years, interval=500)
from IPython.display import HTML
HTML(anim.to_jshtml())

plt.show()
anim.save("deaths_by_cause.gif", writer="pillow", fps=2)
print("Animation saved successfully")

```

```python
# 8) A heatmap of total deaths per state per year i.e., showing "state" on one axis and "year" on the other axis.

pivot_table = df.pivot_table(
    index="state",
    columns="year",
    values="person",
    aggfunc="count",
    fill_value=0)
pivot_table = pivot_table.loc[:, 1980:2016]
top_states = df["state"].value_counts().head(10).index
pivot_table = pivot_table.loc[top_states]


plt.figure(figsize=(16,10))

sns.heatmap(
    pivot_table,
    cmap="Reds",
    linewidths=0.5
)

plt.title("Heatmap of Total Police Deaths per State per Year", fontsize=14)
plt.xlabel("Year")
plt.ylabel("State")
plt.show()

```

```python
# 9) A tree map with three levels: state, city (description), cause

import squarify


# Group data for hierarchy
treemap_df = (
    df.groupby(["state", "dept_name", "cause"])
      .size()
      .reset_index(name="deaths")
)

# Keep top records for readability
treemap_df = treemap_df.sort_values("deaths", ascending=False).head(40)

# Clean cause text
treemap_df["cause"] = treemap_df["cause"].str.replace(
    "Cause of Death:", "", regex=False
)

# Labels for treemap
labels = (
    treemap_df["state"] + "\n" +
    treemap_df["dept_name"] + "\n" +
    treemap_df["cause"]
)

# Plot treemap
plt.figure(figsize=(18, 10))
squarify.plot(
    sizes=treemap_df["deaths"],
    label=labels,
    alpha=0.8
)

plt.title("Treemap of Police Deaths by State, City, and Cause", fontsize=14)
plt.axis("off")
plt.show()
```
