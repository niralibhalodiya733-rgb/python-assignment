```python
#18.Write a Python program to count occurrences of a substring in a string.
text = input("Enter the main string: ")
sub = input("Enter the substring to search: ")

count = 0
index = 0

while index < len(text):
    index = text.find(sub, index)
    if index == -1:
        break
    count += 1
    index += len(sub)

print("Number of occurrences:", count)
```

```python
#29.Write a Python function to get the largest number, smallest num and sum of all from a list.

def list_stats(numbers):
    largest = max(numbers)
    smallest = min(numbers)
    # Use the built-in sum explicitly to avoid conflicts
    total_sum = __builtins__.sum(numbers)
    return largest, smallest, total_sum


nums = [5, 10, 3, 8, 2]
largest, smallest, total = list_stats(nums)

print("Largest number:", largest)
print("Smallest number:", smallest)
print("Sum of all numbers:", total)
```

```python
#63) Write a Python function to check whether a number is perfect or not.
def is_perfect(n):
    """
    Check whether a number is perfect.
    A perfect number is equal to the sum of its proper divisors.
    """
    if n < 1:
        return False

    # Use __builtins__.sum to avoid conflict with any overwritten sum()
    divisors_sum = __builtins__.sum(i for i in range(1, n) if n % i == 0)

    return divisors_sum == n

# Example usage
print(is_perfect(6))   # True
print(is_perfect(28))  # True
print(is_perfect(12))  # False
print(is_perfect(1))   # False
```

```python
#87) When is the finally block executed?
'''The finally block is always executed, regardless of whether an exception occurred or not.
It is often used to release resources, like closing files or network connections.'''

try:
    x = 10 / 0
except ZeroDivisionError:
    print("Error occurred")
finally:
    print("This will always execute")
```

```python
#89) How Do You Handle Exceptions with Try/Except/Finally in Python? Explain with coding snippets.
try:
    x = int(input("Enter a number: "))
    result = 10 / x
except ValueError:
    print("Invalid input. Please enter a number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
finally:
    print("This will always execute.")

'''try → code that may raise exceptions
except → handle specific exceptions
finally → executes no matter what'''
```

```python
#90) Write python program that user to enter only odd numbers, else will raise an exception.
def enter_odd_number():
    try:
        num = int(input("Enter an odd number: "))
        if num % 2 == 0:
            raise ValueError("Even number entered! Only odd numbers allowed.")
        print("You entered:", num)
    except ValueError as e:
        print("Error:", e)

enter_odd_number()
```
